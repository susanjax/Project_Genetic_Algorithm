```python
import pandas as pd
```

```python
original_data = pd.read_csv('/content/combined_with_external_data_purified.csv')
descriptor = pd.read_csv('/content/drive/MyDrive/toxicity_project/data/material_descriptors.csv')
```

```python
# original_data= original_data.drop(['Unnamed: 0'], axis=1)
original_data=original_data.drop(['index'], axis=1)
original_data
```

```python
descriptor = descriptor.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
descriptor
```

combine with material descriptor

```python
original_data['material'] = original_data['material'].replace(['CaHCO3'], 'Ca(HCO3)2')
```

```python
new_one = pd.merge(original_data, descriptor, on='material', how='inner')
```

```python
new_one
```

```python
new_one.isnull().sum().sum()
```

```python
new_one.to_csv('external_with_descriptors.csv')
```
